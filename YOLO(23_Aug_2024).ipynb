```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2
```

```python
model = YOLO('yolov5su.pt')
```

```python
image_path = "/content/drive/MyDrive/YOLO/yolo3.jpeg"
```

```python
img = cv2.imread(image_path)
```

```python
results = model(img)
```

```python
print(type(results))
```

```python
result = results[0]
```

```python
labels = result.names
```

```python
detected_class_indexes = result.boxes.cls.tolist()
```

```python
detected_boxes = result.boxes.xyxy.tolist()
```

```python
detected_labels = [labels[int(index)] for index in detected_class_indexes]
```

```python
print(detected_labels)
```

```python
for bbox, label in zip(detected_boxes, detected_labels):
  x1, y1, x2, y2 = map(int, bbox)
  cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
  cv2.putText(img, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

```python
import cv2
from ultralytics import YOLO
import matplotlib.pyplot as plt

# Load the YOLO model
model = YOLO('yolov5su.pt')

# Path to your video file
video_path = '/content/drive/MyDrive/YOLO/vid1.mp4'

# Open the video file
cap = cv2.VideoCapture(video_path)

# Check if the video opened successfully
if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

# Get the video frame width, height, and FPS
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

# Define the codec and create a VideoWriter object to save the output video
output_path = '/content/drive/MyDrive/YOLO/output_video.mp4'
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec
out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Perform object detection on the frame
    result = model(frame)[0]
    label = result.names
    detected_class_indexes = result.boxes.cls.tolist()
    detected_bboxes = result.boxes.xyxy.tolist()
    detected_labels = [label[int(index)] for index in detected_class_indexes]

    for bbox, label in zip(detected_bboxes, detected_labels):
        x1, y1, x2, y2 = map(int, bbox)
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

    # Write the frame to the output video
    out.write(frame)

# Release the video capture and writer objects
cap.release()
out.release()

print(f"Output video saved to {output_path}")
```
